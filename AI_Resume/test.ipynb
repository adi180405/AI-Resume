```python
import comtypes.client

speaker = comtypes.client.CreateObject("SAPI.SpVoice")

# Get all available voices
voices = speaker.GetVoices()

# List all voices to see what's available
print("Available voices:")
for i in range(voices.Count):
    voice = voices.Item(i)
    print(f"{i}: {voice.GetDescription()}")

# Set to female voice (usually index 1 is female, but check your list)
speaker.Voice = voices.Item(1)  # Change index based on your available voices

# Test it
speaker.Speak("Hello, I am speaking with a female voice now")
```
